```python
def cust_type(x):
    if x <100:
        return "Category A"
    elif (x  > 100) & (x < 199):
        return "Category B"
    else:
        return "Category C"
```

```python
df["Cust_category"] = df["Fare"].apply(cust_type)
```

```python
df[df[["Fare", "Cust_category"]]["Fare"] > 100]
```
